```python
# Verify KG contents
from kg.query import KGQueryEngine
from rdflib import Graph

g = Graph()
g.parse('kg/graphs/mental_health_base.ttl', format='turtle')
engine = KGQueryEngine(g)

conds = engine.get_conditions()
print(f'Total conditions: {len(conds)}')
for uri, label in conds:
    syms = engine.get_symptoms_of_condition(label)
    print(f'{label}: {syms}')
```
